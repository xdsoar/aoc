```python
def sortWord(word):
    wl = list(word)
    wl.sort()
    nword = ''.join(wl)
    return nword
def calcDay8(oldarr):
    symbol1 = ''
    symbol4 = ''
    symbol7 = ''
    symbol8 = ''
    for oldword in oldarr:
        if len(oldword) == 2:
            symbol1 = oldword
        if len(oldword) == 4:
            symbol4 = oldword
        if len(oldword) == 3:
            symbol7 = oldword
        if len(oldword) == 7:
            symbol8 = oldword
    symbol1up = ''
    symbol1down = ''
    symbol6 = ''
    for oldword in oldarr:
        if(len(oldword) != 6):
            continue
        cnts = 0
        for single1 in symbol1:
            if single1 not in oldword:
                symbol6 = oldword
                symbol1up = single1
    for single1 in symbol1:
        if single1 != symbol1up:
            symbol1down = single1
    for oldword in oldarr:
        if(len(oldword) == 5):
            if symbol1up in oldword and symbol1down in oldword:
                symbol3 = oldword
            if symbol1up not in oldword and symbol1down in oldword:
                symbol5 = oldword
            if symbol1up in oldword and symbol1down not in oldword:
                symbol2 = oldword
    single4m = ''
    for single4 in symbol4:
        if single4 in symbol2 and single4 not in symbol1:
            single4m = single4
    for oldword in oldarr:
        if(len(oldword) == 6 and symbol6 != oldword):
            if single4m not in oldword:
                symbol0 = oldword
            else:
                symbol9 = oldword
    arrs = [symbol0, symbol1, symbol2, symbol3, symbol4, symbol5, symbol6, symbol7, symbol8, symbol9]
    
    arrs = list(map(lambda x: sortWord(x), arrs))
    return arrs
        
```

```python
sums = 0
for i in range(len(day8oldarr)):
    day8oldline = day8oldarr[i].split(' ')
    day8newline = day8newarr[i].split(' ')
    day8dict = calcDay8(day8oldline)
    strs = ''
    for word in day8newline:
        orderW = sortWord(word)
        order = day8dict.index(orderW)
        strs = strs + str(order)
    sums = sums + int(strs)
print(sums)
```

```python
#): 3 points.
#]: 57 points.
#}: 1197 points.
#>: 25137 points.
def count10(char):
    if char == ')':
        return 3
    if char == ']':
        return 57
    if char == '}':
        return 1197
    if char =='>':
        return 25137
    return 0

def isMatch(char1, char2):
    if char1 == '(' and char2 == ')':
        return True
    if char1 == '[' and char2 == ']':
        return True
    if char1 == '{' and char2 == '}':
        return True
    if char1 == '<' and char2 == '>':
        return True
    return False
```

```python
stack10 = []
breaklines = []
counts = 0
for line in lines:
    chars = list(line)
    stack10 = []
    for char in chars:
        if char in ['(', '[', '{', '<']:
            stack10.append(char)
        else:
            last = stack10[-1]
            if isMatch(last, char):
                stack10.pop()
            else:
                counts = counts + count10(char)
                breaklines.append(line)
                break
print(counts)
```

```python
#): 1 point.
#]: 2 points.
#}: 3 points.
#>: 4 points
def calScore(char):
    if char == '(':
        return 1
    if char == '[':
        return 2
    if char == '{':
        return 3
    if char == '<':
        return 4
    return 0
def calStack10(stacks):
    counts = 0
    stacks.reverse()
    for char in stacks:
        counts = counts * 5
        counts = counts + calScore(char)
    return counts
```

```python
def increse(cells):
    for line in cells:
        for cell in line:
            cell[0] = cell[0] + 1

def explod(cells, i, j, maxi, maxj):
    if i > 0:
        cells[i-1][j][0] = cells[i-1][j][0] + 1
    if j > 0:
        cells[i][j-1][0] = cells[i][j-1][0] + 1
    if i < maxi - 1:
        cells[i+1][j][0] = cells[i+1][j][0] + 1
    if j < maxj - 1:
        cells[i][j+1][0] = cells[i][j+1][0] + 1
    if i > 0 and j > 0:
        cells[i-1][j-1][0] = cells[i-1][j-1][0] + 1
    if i > 0 and j < maxj - 1:
        cells[i-1][j+1][0] = cells[i-1][j+1][0] + 1
    if i < maxi - 1 and j > 0:
        cells[i+1][j-1][0] = cells[i+1][j-1][0] + 1
    if i < maxi -1 and j < maxj - 1:
        cells[i+1][j+1][0] = cells[i+1][j+1][0] + 1
            
def flashes(cells):
    flashCnt = 0
    for i in range(len(cells)):
        for j in range(len(cells[0])):
            if cells[i][j][0] > 9 and cells[i][j][1] == 0:
                explod(cells, i, j, len(cells), len(cells[0]))
                cells[i][j][1] = 1
                flashCnt = flashCnt + 1
    return flashCnt

def earseall(cells):
    for line in cells:
        for cell in line:
            if cell[0] > 9:
                cell[0] = 0
                cell[1] = 0
def cycleonce():
    tot = 0
    increse(cells)
    cnt = flashes(cells)
    while cnt > 0:
        tot = tot + cnt
        cnt = flashes(cells)
    earseall(cells)
    return tot
def printsp():
    cellsp = []
    for line in cells:
        cellline = []
        for cell in line:
            cellline.append(cell[0])
        cellsp.append(cellline)
    print(cellsp)
```

```python
def isSmall(name):
    return name.lower() == name

def getTarget(edge, cur):
    if edge[0] == cur:
        return edge[1]
    if edge[1] == cur:
        return edge[0]
    return None

def getNext(cur):
    curarray = cur.split('-')
    curnode = curarray[len(curarray)-1]
    visitSmall = list(filter(lambda node: isSmall(node), curarray))
    nextList = list(map(lambda line: getTarget(line, curnode), edge))
    avaliableNext = list(filter(lambda node: node is not None and node not in visitSmall, nextList))
    return avaliableNext
```

```python
minDict = dict()
def getNext(pos, limitx, limity):
    nexts = []
    posd = list(map(lambda x: int(x), pos.split(',')))
    if posd[0] < limitx-1:
        nexts.append((posd[0]+1, posd[1]))
    if posd[1] < limity-1:
        nexts.append((posd[0], posd[1]+1))
    return nexts

def calcos(curpath, cells):
    patharray = curpath.split('->')
    sums = 0
    for pos in patharray[1:]:
        xy = pos.split(',')
        x = int(xy[0])
        y = int(xy[1])
        sums = sums + cells[y][x]
    return sums

def calmins(cells, limitx, limity):
    minDict['0,0'] = 0
    for i in range(limity):
        for j in range(limitx):
            mins = []
            if i == 0 and j == 0:
                mins.append(0)
            if i > 0:
                mins.append(minDict[str(i-1)+','+str(j)] + cells[i][j])
            if j > 0:
                mins.append(minDict[str(i)+','+str(j-1)] + cells[i][j])
            if i < limity - 1 and str(i+1)+','+str(j) in minDict:
                mins.append(minDict[str(i+1)+','+str(j)] + cells[i][j])
            if j < limitx - 1 and str(i)+','+str(j+1) in minDict:
                mins.append(minDict[str(i)+','+str(j+1)] + cells[i][j])
            minDict[str(i)+','+str(j)] = min(mins)
            

def goNext(curpath, cells, limitx, limity):
    patharray = curpath.split('->')
    cur = patharray[len(patharray)-1]
    nexts = getNext(cur, limitx, limity)
    lists = []
    for nextp in nexts:
        nextstr = str(nextp[0]) + ',' + str(nextp[1])
        if nextstr in patharray:
            continue
        ncur = curpath + '->' + nextstr
        ncos = calcos(ncur, cells)
        npcos = minDict.get(nextstr)
        if npcos is not None and npcos < ncos: 
            continue
        else:
            minDict[nextstr] = ncos
        
        if nextp[0] == limitx -1 and nextp[1] == limity -1:
            lists.append(ncur)
        else:
            lists.extend(goNext(ncur, cells, limitx, limity))
    return lists
```

```python
def findrightnum(strin):
    nums = -1
    nume = -1
    for i in range(len(strin)):
        if strin[i] in '0123456789':
            if nums == -1:
                nums = i
            nume = i
        elif nums != -1:
            return (nums, nume)
    return (nums, nume)

def findpairnum(strin):
    nums = []
    numstr = ''
    for i in range(len(strin)):
        if strin[i] in '0123456789':
            numstr = numstr+strin[i]
        elif numstr != '':
            nums.append(int(numstr))
            numstr = ''
        if len(nums) == 2:
            return (nums, i)
    return ([], -1)

def checksplit(strin):
    nums = -1
    nume = -1
    for i in range(len(strin)):
        if strin[i] in '0123456789':
            nume = i
            if nums > -1:
                return (nums, nume)
            else:
                nums = i
        else:
            nums = -1
            nume = -1
    return (nums, nume)

def addup(strin1, strin2):
    return '[' + strin1 + ',' + strin2 + ']'

```

```python
def dual18(strin):
    leftc = []
    num1s = -1
    num1e =-1
    num0s = -1
    num0e = -1
    explodes = -1
    strout = strin
    for i in range(len(strin)):
        s = strin[i]
        if s in '0123456789':
            if num0s == -1:
                num0s = i
                num0e = i
            else:
                num0e = i
        elif num0s != -1:
            num1s = num0s
            num1e = num0e
            num0s = -1
            num0e = -1
        if s == '[':
            leftc.append(s)
        if s == ']':
            leftc = leftc[:-1]
        if len(leftc) > 4:
            explodes = i
            break
    
    if explodes > -1:
        leftp = strin[:explodes]
        middle = strin[explodes:]
        pair = findpairnum(strin[explodes+1:])
        rights = middle[pair[1]+2:]
        rigpos = findrightnum(rights)
        
        if num1s > -1:
            num1 = int(leftp[num1s:num1e+1])
            num1 = num1 + int(pair[0][0])
            leftp = leftp[:num1s] + str(num1) + leftp[num1e+1:]
        
        if rigpos[0] > -1:
            rignum = int(rights[rigpos[0]:rigpos[1]+1])
            nright = int(pair[0][1]) + rignum
            rights = rights[:rigpos[0]] + str(nright) + rights[rigpos[1]+1:]
        result = leftp + '0' + rights
        strout = result
    
    if explodes == -1:
        # checksplit
        splitres = checksplit(strin)
        splits = strin
        if splitres[0] > -1:
            splitnum = int(strin[splitres[0]: splitres[1]+1])
            splitnum1 = splitnum // 2
            splitnum2 = math.ceil(splitnum/2)
            splits = strin[:splitres[0]] + '[' + str(splitnum1) + ',' + str(splitnum2) + ']' + strin[splitres[1]+1:]
            strout = splits
    return strout
    
```

```python
def doreduce(strin):
    strout = dual18(strin)
    while strout != strin:
        strin = strout
        strout = dual18(strin)
    return strout

def dolist18(strlines):
    lines = strlines.split('\n')
    sums = lines[0]
    for i in range(1, len(lines)):
        sums = addup(sums, lines[i])
        sums = doreduce(sums)
    return sums
```

```python
def dosum(sums):
    pattern = '\[([0-9]+),([0-9]+)\]'
    idx = re.search(pattern, sums)
    while idx is not None:
        num1 = int(sums[idx.start(1):idx.end(1)])
        num2 = int(sums[idx.start(2):idx.end(2)])
        num3 = num1*3 + num2*2
        sums = sums[:idx.start()] + str(num3) + sums[idx.end():]
        idx = re.search(pattern, sums)
    return int(sums)
print(dosum('[[1,2],[[3,4],5]]') == 143)
print(dosum('[[[[0,7],4],[[7,8],[6,0]]],[8,1]]') == 1384)
print(dosum('[[[[1,1],[2,2]],[3,3]],[4,4]]') == 445)
print(dosum('[[[[3,0],[5,3]],[4,4]],[5,5]]') == 791)
print(dosum('[[[[5,0],[7,4]],[5,5]],[6,6]]') == 1137)
print(dosum('[[[[8,7],[7,7]],[[8,6],[7,7]]],[[[0,7],[6,6]],[8,7]]]') == 3488)
```
